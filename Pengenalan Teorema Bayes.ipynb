*   ***Conditional probability*** adalah probabilitas terjadinya suatu event jika diketahui event lain.
Probabilitas ini dapat dihitung dengan menggunakan *joint probability*.
<br>
\begin{equation}
P(A|B) = \frac{\displaystyle P(A \cap B)}{\displaystyle P(B)} \tag{1}
\end{equation}

<img src=joint_probability.jpg style="width:400px;height:300px"/>

<b>Ingat!</b> Di dalam *conditional probability* nilai A dan B tidak bersifat simetris.
<br>
\begin{equation}
P(A|B) \neq P(B|A) \tag{2}
\end{equation}
<br>
<br>
Sayangnya *joint probability* ini terkadang tidak dapat langsung dihitung. Cara lain yang dapat digunakan adalah dengan menggunakan persamaan yang ditemukan oleh Rev. Thomas Bayes, yaitu *conditional probability*:
\begin{equation}
P(A|B) = \frac{\displaystyle P(B|A) \times P(A)}{\displaystyle P(B)} \tag{3}
\end{equation}
<img src=thomas_bayes.jpg style="width:400px;height:200px"/>
<br>

<br>
<br>
***Teorema Bayes***
<br>
Teorema bayes adalah cara untuk menghitung *conditional probability* tanpa menggunakan *joint probability*.
<br>
<br>
Ada kasus tertentu di mana nilai $P(B)$ tidak diketahui sehingga dapat dihitung dengan menggunakan rumus berikut:
<br>
<br>
\begin{equation}
P(B) = P(B|A) \times P(A) + P(B|\neg A) \times P(\neg A)\tag{4}
\end{equation}
<br>
<br>
Jika persamaan 4 dimasukkan ke dalam persamaan Bayes (3), maka akan diperoleh:
<br>
<br>
\begin{equation}
P(A|B) = \frac{\displaystyle P(B|A) \times P(A)}{\displaystyle P(B|A) \times P(A) + P(B|\neg A) \times P(\neg A)} \tag{5}
\end{equation}
<br> 
<br>
Perlu kita ingat, jika kita memiliki $P(A)$ maka kita dapat menghitung $P(\neg A)$ sebagai komplemennya dengan cara sebagai berikut:
<br>
<br>
\begin{equation}
P(\neg A) = 1 - P(A) \tag{6}
\end{equation}
<br>
<br>
Sebagai tambahan, jika kita memiliki $P(\neg B|\neg A)$, maka kita dapat menghitung $P(B|\neg A)$ sebagai komplemennya, sebagai contoh:
<br>
<br>
\begin{equation}
P(B|\neg A) = 1 - P(\neg B|\neg A) \tag{7}
\end{equation}
<br>
<br>
**Istilah dalam Teorema Bayes**
<br>
<br>
Setelah penjelasan singkat terkait teorema Bayes di atas, kita akan mencoba untuk memberikan istilah-istilah yang umum digunakan.
<br>
<br>
-  $P(A|B):$ *<b>Posterior</b> probability*
-   $P(A):$ *<b>Prior</b> probability*
-   $P(B|A):$ *Likelihood*
-   $P(B):$ *Evidence*
<br>
<br>
Dari istilah-istilah di atas, maka Teorema Bayes dapat kita tuliskan ulang menjadi persamaan berikut:
<br>
<br>
\begin{equation}
Posterior = \frac{\displaystyle Likelihood \times Prior}{\displaystyle Evidence} \tag{8}
\end{equation}
<br>
<br>
Persamaan (8) di atas dapat kita pahami lebih mudah dengan menggunakan contoh asap dan api. Sebagai contoh, berapakah probabilitas munculnya apa jika diketahui ada asap, di mana $P(Api)$ adalah *Prior*, $P(Asap~|~Api)$ adalah *Likelihood*, dan $P(Asap)$ adalah *Evidence*:
<br>
<br>
\begin{equation}
P(Api~|~Asap) = \frac{\displaystyle P(Asap~|~Api) \times P(Api)}{\displaystyle P(Asap)} \tag{9}
\end{equation}
<br>
<br>
**Contoh 1: Kematian akibat umur tua dan jatuh**
<br>
Jika ada kasus orang yang sudah berumur tua (di atas 80 tahun) terjatuh, berapakah probabilitas orang tua tersebut wafat akibat jatuh? Jika kita asumsikan rata-rata probabilitas orang wafat karena umur tua $P(A)$ adalah 10%, dan probabilitas orang tua jatuh $P(B)$ adalah 5%, dan probabilitas orang tua yang wafat akibat terjatuh $P(B|A)$ adalah 7%. Jika kita masukkan ke dalam teorema Bayes:
<br>
<br>
\begin{equation}
P(A|B) = \frac{\displaystyle P(B|A) \times P(A)}{\displaystyle P(B)}
\tag{10.1}
\end{equation}
<br>
<br>
\begin{equation}
P(Wafat|Jatuh) = \frac{\displaystyle P(Jatuh|Wafat) \times P(Wafat)}{\displaystyle P(Jatuh)}
\tag{10.2}
\end{equation}
<br>
<br>
Jika dituliskan kembali:
<br>
<br>
\begin{equation}
P(Wafat|Jatuh) = \frac{\displaystyle 0.07 \times 0.10}{\displaystyle 0.05}
\tag{11}
\end{equation}
<br>
<br>
Sehingga dapat diperoleh $P(Wafat|Jatuh)=0.14$. Dari hasil ini dapat diambil kesimpulan jika ada seseorang yang sudah berumur tua jatuh maka ada kemungkinan 14% kalau orang tersebut akan wafat. Secara lebih detail, untuk mengaplikasikan perhitungan di atas, kita akan menggunakan bahasa pemrograman Python:

```python
# Hitunglah probabilitas orang berumur tua akibat jatuh
# hitung P(A|B) jika diketahui P(B|A), P(A) and P(B)
def bayes_theorem(p_a, p_b, p_b_given_a):
    # hitung P(A|B) = P(B|A) * P(A) / P(B)
    p_a_given_b = (p_b_given_a * p_a) / p_b
    return p_a_given_b

# P(A)
p_a = 0.10
# P(B)
p_b = 0.05
# P(B|A)
p_b_given_a = 0.07

# hitung P(A|B)
result = bayes_theorem(p_a, p_b, p_b_given_a)
# hasil
print('P(A|B) = %.3f%%' % (result * 100))
```

**Contoh 2: Deteksi Email dan Spam**
<br>
Pernahkah kita mengalami kejadian di mana ada seseorang mengatakan kalau dia sudah mengirimkan email kepada kita tetapi ketika kita cek di dalam email ternyata tidak ada email masuk, lalu akhirnya kita temukan di dalam spam? Jika kita asumsikan probabilitas menerima spam $P(A)$ adalah 2%, probabilitas jika email yang masuk adalah spam dan terdeteksi sebagai spam $P(B|A)$ adalah 99%, dan probabilitas jika sebuah email bukan spam lalu terdeteksi sebagai spam $P(B|\neg A)$ adalah 0.1%, berapakah probabilitas jika email yang terdeteksi sebagai spam adalah spam?
<br>
<br>
\begin{equation}
P(A|B) = \frac{\displaystyle P(B|A) \times P(A)}{\displaystyle P(B)}
\tag{12.1}
\end{equation}
<br>
<br>
\begin{equation}
P(Spam|Terdeteksi) = \frac{\displaystyle P(Terdeteksi|Spam) \times P(Spam)}{\displaystyle P(Terdeteksi)}
\tag{12.2}
\end{equation}
<br>
<br>
Jika dituliskan kembali:
<br>
<br>
\begin{equation}
P(Spam|Terdeteksi) = \frac{\displaystyle 0.99 \times 0.02}{\displaystyle P(Terdeteksi)}
\tag{13}
\end{equation}
<br>
<br>
Karena nilai $P(B)$, di dalam kasus ini sebagai $P(Terdeteksi)$, tidak diketahui, maka kita harus mencari cara untuk menghitungnya dengan menggunakan persamaan (4):
<br>
<br>
\begin{equation}
P(B) = P(B|A) \times P(A) + P(B|\neg A) \times P(\neg A)\tag{14.1}
\end{equation}
<br>
<br>
\begin{equation}
P(Terdeteksi) = P(Terdeteksi|Spam) \times P(Spam) + P(Terdeteksi|\neg Spam) \times P(\neg Spam)\tag{14.1}
\end{equation}
<br>
<br>
Karena kita mengetahui $P(Terdeteksi|\neg Spam)$ adalah 0.1%, maka kita dapat menghitung $P(\neg Spam)$ dengan cara:
<br>
<br>
\begin{equation}
P(\neg Spam) = 1-P(Spam) = 1 - 0.02 = 0.98\tag{15}
\end{equation}
<br>
<br>
Sehingga kita bisa menghitung $P(Terdeteksi)$ sebagai:
<br>
<br>
\begin{equation}
P(\neg Terdeteksi) = 0.99 \times 0.02 + 0.001 \times 0.98 = 0.02078\tag{16}
\end{equation}
<br>
<br>
Dari hasil di atas, dapat kita simpulkan kalau sekitar 2% email yang kita terima akan terdeteksi sebagai spam, meskipun ada kemungkinan email tersebut bukan spam. Dengan hal ini, kita dapat menghitung persamaan (13):
<br>
<br>
\begin{equation}
P(Spam|Terdeteksi) = \frac{\displaystyle 0.99 \times 0.02}{\displaystyle 0.02078} = 0.9528
\tag{13}
\end{equation}
<br>
<br>
Dengan demikian, dari email yang ada di dalam folder Spam, kita dapat yakin kalau 95.2% 

```python
# Menghitung probabilitas email yang ada di dalam folder Spam adalah benar Spam

# Menghitung P(A|B) diketahui P(A), P(B|A), P(B|not A)
def bayes_theorem(p_a, p_b_given_a, p_b_given_not_a):
    # hitung P(not A)
    not_a = 1 - p_a
    # hitung P(B)
    p_b = p_b_given_a * p_a + p_b_given_not_a * not_a
    # hitung P(A|B)
    p_a_given_b = (p_b_given_a * p_a) / p_b
    return p_a_given_b

# P(A)
p_a = 0.02
# P(B|A)
p_b_given_a = 0.99
# P(B|not A)
p_b_given_not_a = 0.001
# hitung P(A|B)
result = bayes_theorem(p_a, p_b_given_a, p_b_given_not_a)
# kesimpulan
print('P(A|B) = %.3f%%' % (result * 100))
```
